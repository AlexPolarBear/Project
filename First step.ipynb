```python
import pandas as pd
import numpy as np
import names as n
```

```python
pd.read_excel('Data1.xlsx')
```

```python
index = np.arange(1, 11)

frame = {'Имя': [n.get_first_name() for i in range(len(index))],
        'Фамилия': [n.get_last_name() for i in range(len(index))],
        'Возраст': [np.random.randint(5,20) for i in range(len(index))]}

frame['№ выступления'] = []
while len(frame['№ выступления']) != len(index):
    for i in range(len(index)):
        k = np.random.randint(i,11)
        if k not in frame['№ выступления'] and k != 0:
            frame['№ выступления'].append(k)
        
frame['Баллы'] = [np.random.randint(20,100) for i in range(len(index))]

frame['Призовое место'] = []
k = 10
p = 11
frame['Баллы'].sort()
for i in range(len(frame['Баллы'])):
    if k <= frame['Баллы'][i]:
        k = frame['Баллы'][i]
        p -= 1
        frame['Призовое место'].append(p)
        
frame['Телефон'] = [("+7" + "(" + str(np.random.randint(100, 1000)) + ")" + str(np.random.randint(100, 1000)) + "-" + str(np.random.randint(10, 100)) + "-" + str(np.random.randint(10, 100))) for i in range(len(index))]

df = pd.DataFrame(frame, index)

df.to_excel('Data2.xlsx')
```
